```python
# 2) Imports and device setup
import torch
from ultralytics import YOLO
import clip
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)
```

```python
# 3) Load models
# YOLOv8n - tiny pretrained detector (uses COCO classes by default)
# If you want the absolute lightest, use 'yolov8n.pt' (this is the default small model).
yolo = YOLO("yolov8n.pt")   # will auto-download if missing

# CLIP model (ViT-B/32) for image-text similarity
clip_model, clip_preprocess = clip.load("ViT-B/32", device=device, jit=False)
clip_model.eval()
```

```python
from google.colab.patches import cv2_imshow
import cv2

def show_boxes(img_bgr, boxes, labels=None, scores=None, title="Image"):
    img = img_bgr.copy()
    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

        txt = ""
        if labels is not None:
            txt += str(labels[i])
        if scores is not None:
            txt += f" {scores[i]:.2f}"

        if txt:
            cv2.putText(img, txt, (x1, max(10, y1 - 5)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    cv2_imshow(img)  # ✅ Works in Colab
```

```python
# 5) Run pipeline: upload or point to an image path, and set a text query
# In Colab, use files.upload() or mount Drive. For local testing, set image_path to your file.
from google.colab import files
uploaded = files.upload()   # choose file
image_path = list(uploaded.keys())[0]

query = "a person selling vegetables"   # change to any scene description
```

```python
# 6) YOLO inference - get candidate boxes
results = yolo.predict(source=image_path, conf=0.25, iou=0.45, max_det=30)  # returns list of Results
res = results[0]   # single image results

# Extract boxes (xyxy), scores, class ids
boxes_xyxy = res.boxes.xyxy.cpu().numpy().astype(int) if len(res.boxes) else np.array([])
scores = res.boxes.conf.cpu().numpy() if len(res.boxes) else []
class_ids = res.boxes.cls.cpu().numpy().astype(int) if len(res.boxes) else []

# Read image for cropping/visual
img_bgr = cv2.imread(image_path)
show_boxes(img_bgr, boxes_xyxy, labels=[int(x) for x in class_ids], scores=scores, title="YOLO Candidate Boxes")
```

```python
# 7) CLIP re-ranking: compute similarity between query and each crop
text_tokens = clip.tokenize([query]).to(device)

best_score = -1.0
best_box = None
all_scores = []

for i, box in enumerate(boxes_xyxy):
    x1,y1,x2,y2 = box
    crop = img_bgr[y1:y2, x1:x2]
    if crop.size == 0:
        all_scores.append(-1.0)
        continue
    crop_pil = Image.fromarray(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)).convert("RGB")
    img_input = clip_preprocess(crop_pil).unsqueeze(0).to(device)

    with torch.no_grad():
        img_feat = clip_model.encode_image(img_input)
        txt_feat = clip_model.encode_text(text_tokens)
        # normalize and cosine similarity
        img_feat = img_feat / img_feat.norm(dim=-1, keepdim=True)
        txt_feat = txt_feat / txt_feat.norm(dim=-1, keepdim=True)
        sim = (img_feat @ txt_feat.T).item()
    all_scores.append(sim)
    if sim > best_score:
        best_score = sim
        best_box = box

print("Best CLIP score:", best_score)
```

```python
# 8) Show best crop and top-k candidates
if best_box is not None:
    x1,y1,x2,y2 = best_box
    best_crop = img_bgr[y1:y2, x1:x2]
    plt.figure(figsize=(6,6)); plt.axis('off'); plt.title("Best Crop (CLIP)"); plt.imshow(cv2.cvtColor(best_crop, cv2.COLOR_BGR2RGB)); plt.show()

# Optionally show top-3 crops
topk = 3
idxs = np.argsort(all_scores)[-topk:][::-1]
crops = []
for idx in idxs:
    x1,y1,x2,y2 = boxes_xyxy[idx]
    crop = img_bgr[y1:y2, x1:x2]
    crops.append(crop)
    plt.figure(figsize=(4,4)); plt.axis('off'); plt.title(f"Rank {list(idxs).index(idx)+1} score={all_scores[idx]:.3f}"); plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)); plt.show()
```
